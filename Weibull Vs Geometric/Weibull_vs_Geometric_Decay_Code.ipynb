```python
import numpy as np
import matplotlib.pyplot as plt

# Data
weeks = np.arange(0, 15, 1)  # X-axis points from 0 to 14 by 1
impressions = [1, 1, 1, 100, 50, 25, 13, 7, 4, 2, 1, 1, 1, 1, 1]  # Y-axis points (15 points to match weeks)

# Create figure and axis with specific size
plt.figure(figsize=(12, 6))

# Set the grid style
plt.grid(True, linestyle='--', alpha=0.7)

# Plot the main line
plt.plot(weeks, impressions, '-o', color='#ffa500', label='Geometric Advertising Adstock', linewidth=2, markersize=6)

# Create and style the advertisement period rectangle
plt.axvspan(2, 3, alpha=0.2, color='blue', label='Impressions')

# Customize the plot
plt.title('Geometric Distribution of Advertising Impressions Over Time', pad=15, fontsize=12)
plt.xlabel('Weeks')
plt.ylabel('Impressions/Adstock')

# Set axis limits
plt.xlim(0, 14)
plt.ylim(0, 110)

# Add legend
plt.legend(loc='upper right')

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Save the plot as a PNG file
plt.savefig("Geometric Adstock.png", dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
weeks = np.linspace(0, 20, 17)  # Create 17 evenly spaced points from 0 to 20
impressions = [0.01, 0.01, 0.01, 0.95, 1.7, 2.1, 2.1, 1.85, 1.4, 1.0, 0.6, 0.35, 0.2, 0.1, 0.05, 0.02, 0.01]  # 17 points

# Create figure and axis with specific size
plt.figure(figsize=(12, 6))

# Set the grid style
plt.grid(True, linestyle='--', alpha=0.7)

# Plot the main line
plt.plot(weeks, impressions, '-o', color='#800080', label='Weibull Advertising Adstock', linewidth=2, markersize=6)

# Create and style the advertisement period rectangle
plt.axvspan(1, 2.5, alpha=0.2, color='blue', label='Impressions')

# Customize the plot
plt.title('Weibull Adstock Over Time', pad=15, fontsize=12)
plt.xlabel('Weeks')
plt.ylabel('Impressions')

# Set axis limits
plt.xlim(0, 20)
plt.ylim(0, 2.5)

# Add legend
plt.legend(loc='upper right')

# Adjust layout to prevent label cutoff
plt.tight_layout()


# Save the plot as a PNG file
plt.savefig("Weibull Adstock.png", dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```
